## 📂 Part 1: File Handling (45 min)

### How to open a file — the `open()` function

| Mode | Symbol | What it does |
|------|--------|--------------|
| Read | `'r'` | 📖 Read an existing file |
| Write | `'w'` | ✏️ Write to file (creates new / **overwrites**) |
| Append | `'a'` | ➕ Add to end of file (does NOT overwrite) |
| Create | `'x'` | 🆕 Create file (fails if it already exists) |

### The Golden Rule 🥇
Always use `with open(...)` — it automatically closes the file for you!

````python
# ✏️ WRITE — Create a file and put text into it
with open("notes.txt", "w") as f:
    f.write("Hello Students!\n")
    f.write("Welcome to Python Class 4!\n")

print("✅ File created successfully!")
````

````python
# 📖 READ — Read back what we wrote
with open("notes.txt", "r") as f:
    content = f.read()

print("📄 File contents:")
print(content)
````

````python
# ➕ APPEND — Add more without deleting old content
with open("notes.txt", "a") as f:
    f.write("Today we learned File Handling! 🎉\n")

# Read the updated file
with open("notes.txt") as f:
    print(f.read())
````

## 📊 Part 2: CSV Files (30 min)

### What is CSV?
**CSV = Comma Separated Values**

```
Name, Age, Marks
Ali,  20,  95
Sara, 21,  92
```

- Opens in **Excel / Google Sheets** 📊
- Used everywhere: bank statements, student records, sales data
- Python has a built-in `csv` module — no install needed!

````python
import csv

# ✏️ Writing a CSV file
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age", "Marks"])  # Header row
    writer.writerow(["Ali",  20,    95])
    writer.writerow(["Sara", 21,    92])
    writer.writerow(["Ahmed",19,    88])
    writer.writerow(["Fatima",22,   97])

print("✅ students.csv created!")
````

````python
# 📖 Reading a CSV file
with open("students.csv") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
````

````python
# 🌟 Even better — DictReader (reads as dictionary!)
with open("students.csv") as f:
    reader = csv.DictReader(f)
    for student in reader:
        print(f"👤 {student['Name']} | Age: {student['Age']} | Marks: {student['Marks']}")
````

## 🌐 Part 3: JSON (30 min)

### What is JSON?
**JSON = JavaScript Object Notation**

- The language of the **internet** 🌍
- When WhatsApp sends your message, it travels as JSON
- When you log into an app, the server responds with JSON
- Looks almost exactly like a Python **dictionary**!

```json
{
    "name": "Ali",
    "age": 20,
    "marks": 95
}
```

### CSV vs JSON — When to use which?
| | CSV | JSON |
|--|-----|------|
| **Best for** | Tables, spreadsheets | Complex / nested data |
| **Looks like** | Excel | Python dictionary |
| **Used in** | Reports, exports | APIs, web apps |

````python
import json

# 💾 Save a dictionary to JSON file
student = {
    "name": "Ali",
    "age": 20,
    "marks": 95,
    "subjects": ["Python", "Maths", "English"],  # Lists work too!
    "passed": True
}

with open("student.json", "w") as f:
    json.dump(student, f, indent=4)   # indent=4 makes it pretty!

print("✅ JSON file saved!")
````

````python
# 📖 Load JSON back into Python
with open("student.json") as f:
    data = json.load(f)

print("Loaded data:", data)
print(f"\nStudent name: {data['name']}")
print(f"Subjects: {', '.join(data['subjects'])}")
````

````python
# 🔄 JSON in strings (like what an API sends you)
json_string = '{"city": "Karachi", "temp": 35}'

weather = json.loads(json_string)   # loads = load from STRING
print(f"City: {weather['city']}, Temperature: {weather['temp']}°C")

# Turn dict back into string
back_to_string = json.dumps(weather)
print(f"As string: {back_to_string}")
````

### 🗒️ JSON Cheat Sheet
| Function | Use |
|----------|-----|
| `json.dump(data, file)` | Save dict → file |
| `json.load(file)` | Load file → dict |
| `json.dumps(data)` | Dict → string |
| `json.loads(string)` | String → dict |

## 🛡️ Part 4: Error Handling (30 min)

### Why do programs crash? 💥
- User types `"hello"` when you expected a number
- File doesn't exist
- Dividing by zero
- Internet connection drops

**Professional programs NEVER just crash.** They handle errors gracefully!

### The `try / except` structure
```python
try:
    # 🎯 Code that might fail
except SomeError:
    # 🛡️ What to do if it fails
else:
    # ✅ Runs only if NO error happened
finally:
    # 🔄 ALWAYS runs, error or not
```

````python
# 🛡️ Basic error handling
try:
    number = int(input("Enter a number: "))
    result = 100 / number
    print(f"100 ÷ {number} = {result}")
except ValueError:
    print("❌ Please enter a NUMBER, not text!")
except ZeroDivisionError:
    print("❌ You can't divide by zero!")
else:
    print("✅ Calculation successful!")
finally:
    print("🏁 Program finished.")
````

````python
# 📂 Error handling with files
try:
    with open("missing_file.txt") as f:
        print(f.read())
except FileNotFoundError:
    print("❌ Oops! That file doesn't exist.")
    print("💡 Let me create it for you...")
    with open("missing_file.txt", "w") as f:
        f.write("File created as backup!")
    print("✅ File created!")
````

### 🎮 Guess the Error Game!
Look at each code snippet and guess the error before running it.

| Code | Your Guess | Actual Error |
|------|------------|------|
| `print(age)` | ? | `NameError` |
| `int('Hello')` | ? | `ValueError` |
| `10 / 0` | ? | `ZeroDivisionError` |
| `open('ghost.txt', 'r')` | ? | `FileNotFoundError` |
| `[1,2,3][10]` | ? | `IndexError` |

````python
# Test each one in a try/except to see the error message!
errors_to_test = [
    ("print(age)",          lambda: print(age)),
    ("int('Hello')",        lambda: int('Hello')),
    ("10 / 0",              lambda: 10/0),
    ("open missing file",   lambda: open('ghost.txt', 'r')),
    ("[1,2,3][10]",         lambda: [1,2,3][10]),
]

for description, code in errors_to_test:
    try:
        code()
    except Exception as e:
        print(f"🔴 {description} → {type(e).__name__}: {e}")
````

## 🚀 Mini Project — Student Record System (20 min)

We'll build a menu-driven program that:
- ➕ Add a student
- 👀 View all students
- 💾 Save as CSV
- 🌐 Save as JSON
- 🚪 Exit

**With full error handling!**

````python
import csv
import json

students = []  # Our in-memory list

def add_student():
    name = input("Student name: ").strip()
    try:
        age = int(input("Age: "))
        marks = float(input("Marks (out of 100): "))
    except ValueError:
        print("❌ Age and Marks must be numbers! Student not added.")
        return
    students.append({"name": name, "age": age, "marks": marks})
    print(f"✅ {name} added!")

def view_students():
    if not students:
        print("📭 No students yet!")
        return
    print(f"\n{'Name':<15} {'Age':<6} {'Marks'}")
    print("-" * 30)
    for s in students:
        print(f"{s['name']:<15} {s['age']:<6} {s['marks']}")

def save_csv():
    if not students:
        print("❌ Nothing to save!")
        return
    with open("records.csv", "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=["name", "age", "marks"])
        writer.writeheader()
        writer.writerows(students)
    print("✅ Saved to records.csv")

def save_json():
    if not students:
        print("❌ Nothing to save!")
        return
    with open("records.json", "w") as f:
        json.dump(students, f, indent=4)
    print("✅ Saved to records.json")

# 🎮 Main Menu Loop
while True:
    print("\n" + "="*35)
    print("   🎓 Student Record System")
    print("="*35)
    print("1. ➕ Add Student")
    print("2. 👀 View Students")
    print("3. 💾 Save as CSV")
    print("4. 🌐 Save as JSON")
    print("5. 🚪 Exit")
    print("="*35)

    choice = input("Choose (1-5): ").strip()

    if   choice == "1": add_student()
    elif choice == "2": view_students()
    elif choice == "3": save_csv()
    elif choice == "4": save_json()
    elif choice == "5":
        print("👋 Goodbye! See you next class!")
        break
    else:
        print("❓ Invalid choice. Pick 1-5.")
````
