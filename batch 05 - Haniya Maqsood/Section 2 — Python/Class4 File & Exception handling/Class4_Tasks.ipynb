# 📂 PART 1 — File Handling
### How to open a file
| Mode | What it does |
|------|--------------|
| `'w'` | Write — creates new file or **overwrites** existing |
| `'r'` | Read — reads an existing file |
| `'a'` | Append — adds to the end without deleting old content |

**Golden Rule 🥇** — Always use `with open(...)`. It closes the file automatically.

### ✅ Example 1 — Write, Append & Read

````python
# STEP 1 — Write two lines to a file
with open("shop.txt", "w") as f:
    f.write("Shop Name: Ali's Store\n")
    f.write("City: Karachi\n")

# STEP 2 — Append one more line
with open("shop.txt", "a") as f:
    f.write("Owner: Ali Khan\n")

# STEP 3 — Read and print everything
with open("shop.txt", "r") as f:
    content = f.read()

print("--- shop.txt ---")
print(content)
````

### ✅ Example 2 — Reading Line by Line

````python
# First, create a file with some items
with open("groceries.txt", "w") as f:
    f.write("Milk\nBread\nEggs\nButter\nJuice\n")

# Read it back LINE BY LINE using readlines()
with open("groceries.txt", "r") as f:
    lines = f.readlines()    # returns a list — one item per line

print(f"You have {len(lines)} items to buy:")
for i, item in enumerate(lines, start=1):
    print(f"  {i}. {item.strip()}")    # .strip() removes the \n at the end
````

# 📊 PART 2 — CSV Files

**CSV = Comma Separated Values** — a simple format that opens in Excel.

```
Name,Age,Marks
Ali,20,95
Sara,21,92
```

Python has a built-in `csv` module — no installation needed!

### ✅ Example 3 — Write & Read a CSV

````python
import csv

# --- WRITE ---
with open("products.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Product", "Category", "Price"])   # header row
    writer.writerow(["Laptop",  "Electronics", 85000])
    writer.writerow(["Chair",   "Furniture",   12000])
    writer.writerow(["Notebook","Stationery",  150])

print("✅ products.csv created!")

# --- READ using DictReader (reads each row as a dictionary) ---
print("\n--- Product List ---")
with open("products.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"  {row['Product']:<12} | {row['Category']:<12} | Rs.{row['Price']}")
````

# 🌐 PART 3 — JSON Files

JSON is perfect for storing **complex, structured data** — like a dictionary with lists inside.

| Function | Does what |
|----------|-----------|
| `json.dump(data, file)` | Save a dictionary → file |
| `json.load(file)` | Load a file → dictionary |


### ✅ Example 5 — Save and Load a Dictionary as JSON

````python
import json

# A dictionary — just like what you have in memory
student = {
    "name"    : "Ali",
    "age"     : 20,
    "marks"   : 92.5,
    "subjects": ["Python", "Math", "English"]
}

# --- SAVE to file ---
with open("student.json", "w") as f:
    json.dump(student, f, indent=4)   # indent=4 makes it human-readable

print("✅ Saved to student.json")

# --- LOAD back from file ---
with open("student.json") as f:
    loaded = json.load(f)

print(f"\nName    : {loaded['name']}")
print(f"Age     : {loaded['age']}")
print(f"Marks   : {loaded['marks']}")
print(f"Subjects: {', '.join(loaded['subjects'])}")
````

### ✅ Example 6 — A List of Dictionaries in JSON

````python
import json

# A list of dictionaries (like a table with rows)
inventory = [
    {"item": "Pen",     "qty": 100, "price": 20},
    {"item": "Notebook","qty": 50,  "price": 120},
    {"item": "Ruler",   "qty": 30,  "price": 35},
]

# Save the whole list
with open("inventory.json", "w") as f:
    json.dump(inventory, f, indent=4)

# Load and display
with open("inventory.json") as f:
    loaded_inv = json.load(f)

print(f"{'Item':<12} {'Qty':<6} {'Price':<8} {'Total Value'}")
print("-" * 40)
for product in loaded_inv:
    total = product['qty'] * product['price']
    print(f"{product['item']:<12} {product['qty']:<6} Rs.{product['price']:<5} Rs.{total}")
````
