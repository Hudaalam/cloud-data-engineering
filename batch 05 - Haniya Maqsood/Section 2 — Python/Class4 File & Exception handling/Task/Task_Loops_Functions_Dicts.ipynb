## ✅ Example — School Report System

Study this example carefully before solving the task below.

````python
# ── Data: a list of student dictionaries ──────────────────────
students = [
    {"name": "Ali",   "marks": {"Math": 80, "English": 72, "Science": 90}},
    {"name": "Sara",  "marks": {"Math": 95, "English": 88, "Science": 76}},
    {"name": "Ahmed", "marks": {"Math": 60, "English": 55, "Science": 65}},
]

# ── Function 1: calculate average of a student's marks ────────
def get_average(marks_dict):
    total = 0
    for subject, score in marks_dict.items():   # loop over dictionary
        total += score
    return round(total / len(marks_dict), 1)

# ── Function 2: decide grade based on average ─────────────────
def get_grade(average):
    if average >= 80:
        return "A"
    elif average >= 65:
        return "B"
    else:
        return "C"

# ── Function 3: print the full report ─────────────────────────
def print_report(students):
    print(f"{'Name':<10} {'Average':<10} {'Grade'}")
    print("-" * 30)
    for student in students:               # loop over the list
        avg   = get_average(student["marks"])
        grade = get_grade(avg)
        print(f"{student['name']:<10} {avg:<10} {grade}")

print_report(students)
````
